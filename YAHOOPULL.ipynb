```python
stocks = pd.read_csv('../data/tickers.csv')
```

```python
stocks.head()
stocks = stocks[['Symbol', 'IPO Year', 'Sector', 'Industry']]
```

```python
tickers = stocks[['Symbol']]
```

```python
ticker_names = tickers['Symbol'].unique()
```

```python

all_tickers = []


for ticker in range(1, len(ticker_names)):
    print(ticker_names[ticker])
    
    if '/' in ticker_names[ticker]:
        pass
    else:
        try:
            tick = yf.download(ticker_names[ticker], 
                                  start='2010-01-01', 
                                  end='2020-09-30', 
                                  progress=True) 

            tick = tick.T
            tick['name'] = ticker_names[ticker]
            all_tickers.append(tick)
        except:
            print("Whoops")
    

```

```python
large = pd.concat(all_tickers)
## We either need to do a PCA or high dimensional clustering 
## PCA might capture major market movements 

### Then we should segment based on industry etc - or do a clustering on industry
```

```python
large['name'].head()
```

```python
large.to_csv('../data/large.csv')
```
