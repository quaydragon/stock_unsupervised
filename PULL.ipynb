```python
stocks = pd.read_csv('../data/tickers.csv')
```

```python
stocks.head()
stocks = stocks[['Symbol', 'IPO Year', 'Sector', 'Industry']]
```

```python
tickers = stocks[['Symbol']]
```

```python
ticker_names = tickers['Symbol'].unique()
```

```python
all_tickers = []



for ticker in range(1, 100):
    print(ticker_names[ticker])
    
    if '/' in ticker_names[ticker]:
        pass
    else:
        try:
            tick = yf.download(ticker_names[ticker], 
                                  start='2010-01-01', 
                                  end='2020-09-30', 
                                  progress=True) 

            tick = tick.T
            tick['name'] = ticker_names[ticker]
            all_tickers.append(tick)
        except:
            print("Whoops")
    

large = pd.concat(all_tickers)
```
